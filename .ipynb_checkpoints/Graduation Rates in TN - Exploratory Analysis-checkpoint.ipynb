```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('white')
sns.despine()
sns.set(font_scale=1.2)
plt.subplots_adjust(hspace=0.5)
plt.tight_layout()
```

```python
data = pd.read_csv('./data/grad_demo_fin.csv')
```

```python
data
```

```python
data.info()
```

```python
figure, axes = plt.subplots(3,6, figsize=(30,13))

for ax, feature in zip(axes.flat, data.columns):
    ax.grid(axis='y', color='black',alpha=.3)
    ax.grid(False, axis='x')    
    g = sns.histplot(data[feature],ax=ax, bins=30, kde=True, color='purple', line_kws={'linestyle': '--','lw':2})
    g.set(ylabel=None)
```

```python
data.corr()
```

```python
top_correlated = data.corr().iloc[0].sort_values(ascending=False).index[1:]
```

```python
figure, axes = plt.subplots(3,6, figsize=(30,13), sharey=True)

for ax, feature in zip(axes.flat, top_correlated):
    ax.grid(axis='y', color='black',alpha=.3)
    ax.grid(False, axis='x')    
    g = sns.regplot(data=data,x=feature,y='grad_rate',ax=ax, line_kws= {'color': 'cyan'}, color='purple')
```

#### Observations
An interesting trend here is that the more a school spends per pupil, the lower the graduation rate seems to be. This seems to be counter intuitive, but it would make sense if a struggling school receives more funds in hopes up driving up the graduation rate. Another noteable observation is that the higher proportion a school has of African Americans and Economically disadvantaged individuals, the lower the graduation rate seems to be.
