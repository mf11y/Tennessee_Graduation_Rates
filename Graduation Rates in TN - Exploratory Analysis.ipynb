```python
def create_regplots(y_feature):
    r_label = data.corr().loc[y_feature].sort_values(ascending=False)[1:].values 
    top_correlated = data.corr().loc[y_feature].sort_values(ascending=False).index[1:]
    
    figure, axes = plt.subplots(3,5, figsize=(25,13))
    
    for ax, x_feature,r_coef in zip(axes.flat, top_correlated,r_label):
        ax.grid(axis='y', color='black',alpha=.3)
        ax.grid(False, axis='x')
        sns.regplot(data=data,x=x_feature,y=y_feature,ax=ax,
                        line_kws={'color': 'cyan'}, color='purple',
                        label = 'r^2:  ' + str(round((r_coef*r_coef), 2)))
        ax.legend(loc='best')
        
    plt.tight_layout()
```

```python
create_regplots('grad_rate')
```

#### Observations
An interesting trend here is that the more a school spends per pupil, the lower the graduation rate seems to be. This seems to be counter intuitive, but it would make sense if a struggling school receives more funds in hopes up driving up the graduation rate. Another noteable observation is that the higher proportion a school has of African Americans and Economically disadvantaged individuals, the lower the graduation rate seems to be. This could closely tie in with being economically disadvantaged.

```python
create_regplots('ECONOMICALLY_DISADVANTAGED')
```
