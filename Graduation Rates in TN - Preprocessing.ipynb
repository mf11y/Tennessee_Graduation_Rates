```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('./Data/grad_demo_fin.csv')
```

```python
data
```

```python
data.info()
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(data.drop(columns=['grad_rate']))
scaled_data = pd.DataFrame(scaled_data, columns=data.drop(columns=['grad_rate']).columns)
scaled_data = pd.concat([scaled_data,data[['grad_rate']]], axis=1)
```

```python
scaled_data.to_csv('./Data/scaled_grad_demo_fin.csv', index=False)
```
